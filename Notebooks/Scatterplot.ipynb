```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
df=pd.read_csv("C:/Users/rauna/Desktop/IDS Project/credit_train_clean_sample_no_outliers2.csv")
plt.figure(figsize=(20,10))
plt.plot("Number of Open Accounts", "Maximum Open Credit", data=df,linestyle='',marker='o',markersize=2,color='blue')
ax=plt.gca()
ax.set_facecolor("black")
plt.xlabel("Number of Open Accounts")
plt.ylabel("Maximum Open Credit")
r,p=pearsonr(df["Number of Open Accounts"],df["Maximum Open Credit"])
print(r)
```
